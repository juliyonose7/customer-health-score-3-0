# 02 - EDA
Explore transactional and service datasets.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

root = Path.cwd().parents[0]
raw = root / 'data' / 'raw'
tx = pd.read_csv(raw / 'transactions.csv', parse_dates=['date'])
tickets = pd.read_csv(raw / 'support_tickets.csv', parse_dates=['date'])
social = pd.read_csv(raw / 'social_mentions.csv', parse_dates=['date'])
tx.head()
```

```python
print('Transactions:', tx.shape)
print('Support tickets:', tickets.shape)
print('Social mentions:', social.shape)
tx['amount'].describe()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(tx['amount'], bins=40, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()
```
